```python
#1- Write a program that takes a list of numbers from the user (until they enter ‘done’),then prints the list,its sorted version (ascending and descending) :

my_list = []
while True:
    x = input("Enter number: ")
    if x == "done":
        break
    my_list.append(float(x))

print(my_list)
print(sorted(my_list))
print(sorted(my_list, reverse=True))
```

```python
# 2- Create a function modify_list(lst) that receives a list, appends 100 to it,
# inserts 999 at index 2, removes the first occurrence of 50 (if exists), pops
# the last element, and returns the modified list.

def modify_list(lst):
    lst.append(100)
    lst.insert(2, 999)
    if 50 in lst:
        lst.remove(50)
    lst.pop()
    return lst
```

```python
# 3- Write a program that asks the user for 5 student names and their grades,
# stores them in a list of lists (e.g. [[‘Ali’, 85], [‘Sara’, 92]]),
# then prints the student with the highest and lowest grade.

students = []
for _ in range(5):
    name = input("Name: ")
    grade = int(input("Grade: "))
    students.append([name, grade])

highest = max(students, key=lambda x: x[1])
lowest = min(students, key=lambda x: x[1])

print("Highest:", highest)
print("Lowest:", lowest)
```

```python
#5- Write a program that asks the user for their name and age (two inputs), stores them in a small list [name, age], then uses if-elif-else to print different messages: “Young” if age < 18, “Adult” if 18–60, “Senior” otherwise :

name = input("Name: ")
age = int(input("Age: "))

if age < 18:
    print("Young")
elif 18 <= age <= 60:
    print("Adult")
else:
    print("Senior")
```

```python
# 6- Implement a program that creates a list of temperatures (you choose 10
# values),then uses a for loop to count how many are above 30 and how many below
# 15, and prints both counts.  7- Create a function
# remove_all_occurrences(lst, value) that removes all occurrences of a given
# value from the list.

my_list = [12, 35, 28, 40, 10, 18, 33, 14, 22, 45,90,5,7,34]
above30 = 0
below15 = 0

for i in my_list:
    if i > 30:
        above30 += 1
    if i < 15:
        below15 += 1

print("Above 30 :", above30)
print("Below 15 :", below15)
```

```python
#7- Create a function remove_all_occurrences(lst, value) that removes all occurrences of a given value from the list :

def remove_all_occurrences(lst, value):
    while value in lst:
        lst.remove(value)
    return lst
```

```python
#8- Write a number guessing game: computer has a fixed number (e.g. 42), user guesses via input() in a while loop, give hints “too high” / “too low”, count attempts, end when correct :

x = 42
count = 0

while True:
    guess = int(input("Guess: "))
    count += 1
    if guess < x:
        print("Too low")
    elif guess > x:
        print("Too high")
    else:
        print("Correct Attempts:", count)
        break
```

```python
#9- Create a shopping list program using a list. In a while True loop, show a menu: .Add item .Remove item by name (use remove() if found, else message) .Show sorted list .Exit Use input() for choices and item :

shopping_list = ["milk","bread","eggs","rice","chicken","apples","cheese"]

while True:
    print("1-Add  2-Remove  3-Show sorted  4-Exit")
    choice = input("Choice: ")

    if choice == "1":
        item = input("Item: ")
        shopping_list.append(item)

    elif choice == "2":
        item = input("Remove item: ")
        if item in shopping_list:
            shopping_list.remove(item)
        else:
            print("Not found")

    elif choice == "3":
        print(sorted(shopping_list))

    elif choice == "4":
        break
```

```python
#10- Write a program that asks the user to enter 7 city names, stores them in a list, then uses a for loop to print each city with its length (e.g. “Cairo – 5 letters”) :

city_list = []
for i in range(7):
    city = input("City: ")
    city_list.append(city)

for c in city_list:
    print(c, "-", len(c), "letters")
```
